```python
from numpy import *
from matplotlib.pyplot import *
import matplotlib.animation
```

```python
width=100
height=100
initProb=0.4
```

```python
def inicializa():
  global time,config, nextConfig
  time=0
  config=zeros([height,width])
  for x in range(width):
    for y in range(height):
      if random.random() < initProb:
        state = 1
      else:
        state = 0
      config[y,x] = state
  nextConfig=zeros([height,width])
```

```python
def observa():
  fig,ax=subplots()
  im=ax.imshow(config,vmin=0,vmax=1,cmap='binary',animated=True)
  axis('image')
  title('t='+str(time))
  show()
```

```python
def actualiza():
  global time, config, nextConfig,state

  time += 1

  for x in range(width):
    for y in range(height):
      state = config[y,x]
      vivos=0
      for dx in range(-1,2):
        for dy in range(-1,2):
          vivos+= config[(y+dy)%height,(x+dx)%width]
      if state == 0 and vivos == 3:
        state = 1
      elif state == 1 and (vivos < 3 or vivos > 4):
        state=0
      nextConfig[y,x]=state
    config,nextConfig = nextConfig, config
```

```python
%matplotlib inline
inicializa()
observa()
for i in range(50):
  actualiza()
  observa()
```

```python
%matplotlib notebook
ani = matplotlib.animation.ArtistAnimation(fig,ims,interval=50,blit=True,
                                repeat_delay=10000)
show()
```
