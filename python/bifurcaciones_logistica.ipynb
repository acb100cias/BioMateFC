```python
from numpy import *
from matplotlib.pyplot import *
```

```python
def f(r,x):
    return r*x*(1-x)
```

```python
n = 10000
r = linspace(1, 3.925, n)
iterations = 1000
last = 100
x = 1e-5 * np.ones(n)
lyapunov = np.zeros(n)
fig,ax1  = subplots(1, 1, figsize=(20, 10))
for i in range(iterations):
    x = f(r, x)
    lyapunov += np.log(abs(r - 2 * r * x))
    if i >= (iterations - last):
        ax1.plot(r, x, ',k', alpha=.25)
ax1.set_xlim(1., 3.925)
ax1.set_title('Diagrama de bifurcaciones')
xlabel('r')
ylabel('X')
```
