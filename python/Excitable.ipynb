```python
from numpy import *
from matplotlib.pyplot import*
```

```python
uc=0.25
def f(u):
    return u*(u-uc)*(1-u)
```

```python
f(0.27)
```

```python
U=arange(-0.2,1.1,0.01)
```

```python
U
```

```python
plot(U,f(U))
scatter([0,uc,1],[0,f(uc),f(1)],color='r')
axhline(y=0, color='k')
axvline(x=0, color='k')
```

```python
uc=0.25
def fv(u,v):
    return u*(u-uc)*(1-u)-v
```

```python
for i in arange(0,0.14,0.032):
    plot(U,fv(U,i),label=i)
    #figsize=(20,20)
    axhline(y=0, color='k')
    axvline(x=0, color='k')
    legend()
```

```python
from scipy.integrate import odeint
```

```python
epsilon=0.1
k=1.
def F(U,t):
    u,v=U
    return array([u*(u-uc)*(1-u)-v,epsilon*(k*u-v)])
```

```python
odeint(F,[0,-0.2],np.linspace(-1, 100, 100)).T[1]
```

```python
np.linspace(-1, 100, 100)
```

```python
plot(odeint(F,[0,-0.2],np.linspace(-1, 100, 100)).T[0],label='u(t)')
plot(odeint(F,[0,-0.2],np.linspace(-1, 100, 100)).T[1],label='v(t)')
legend()
```

```python
plot(odeint(F,[0,-0.2],np.linspace(-1, 1000, 3000)).T[1],odeint(F,[0,-0.2],np.linspace(-1, 1000, 3000)).T[0])
```

```python
for j in arange(0,1,0.1):
    plot(odeint(F,[0+j,-0.2+j],np.linspace(-1, 1000, 3000)).T[1],odeint(F,[0+j,-0.2+j],np.linspace(-1, 1000, 3000)).T[0],label=j)
legend()
```
