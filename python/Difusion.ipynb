```python
from numpy import *
from matplotlib.pyplot import *
from random import *
```

```python
dt=0.002;
Dz=1;
dx=1;
dy=1;
nx=120;
ny=120;
nt=50000;
p=zeros([nx,ny,nt]);
```

```python
for f in range(20):
    p[int(round((nx-1)*random())),int(round((ny-1)*random())),0]=sign(random()-0.51)

```

```python
for m in range(1,nt):
    p[1:nx-1,1:ny-1,m]=p[1:nx-1,1:ny-1,m-1]+dt*Dz*((p[2:nx,1:ny-1,m-1]-2*p[1:nx-1,1:ny-1,m-1]+p[0:nx-2,1:ny-1,m-1])/np.power(dx,2)+(p[1:nx-1,2:ny,m-1]-2*p[1:nx-1,1:ny-1,m-1]+p[1:nx-1,0:ny-2,m-1])/np.power(dy,2));
```

```python
fig = figure()
for g in range(18):
    subplot(3,6,g+1)
    axis([0,nx,0,ny])
    pcolor(p[:,:,int(round(g*(nt-1)/17))],vmin=-0.001,vmax=0.001,cmap='jet')
    axis('off')
```
