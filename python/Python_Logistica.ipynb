```python
import numpy as np 
import scipy.integrate as sc
import matplotlib.pyplot as plt 
```

```python
K=0.02
def f(x,t):
    return K*x
```

```python
T=np.arange(0,100,0.001)
```

```python
T
```

```python

X=sc.odeint(f,0.01,T)
```

```python
plt.plot(T,X)
```

```python
for ci in [0.1,0.5,0.6,1]:
    X=sc.odeint(f,ci,T)
    plt.plot(T,X)
```

```python
K=20
r=0.02
def logistica(x,t):
    return x*r*(1-x/K)
```

```python
T1=np.arange(0,400,0.001)
X1=sc.odeint(logistica,1,T1)
X2=sc.odeint(logistica,20.,T1)
plt.plot(T1,X1)
plt.plot(T1,X2)
```

```python
L=[5,10,15,20,22]
for ci in L:
    X1=sc.odeint(logistica,ci,T1)
    plt.plot(T1,X1)
```

```python
K=20
r=0.02
E=0.03
def logisticaC(x,t):
    return x*r*(1-x/K)-E*x
```

```python
L=[5,10,15,20,22]
for ci in L:
    X1=sc.odeint(logisticaC,ci,T1)
    plt.plot(T1,X1)
```
