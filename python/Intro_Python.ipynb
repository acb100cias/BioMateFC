```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import *
```

```python
def LV(X,t,a,b,c,d):
  x,y=X
  dx=a*x-b*x*y
  dy=-c*y+d*x*y
  return dx , dy
```

```python
Y=odeint(LV,(0.1,0.1),T,args=(0.5,0.3,0.5,0.4))
```

```python
T=arange(0,100,0.01)
```

```python
plot(T,Y[:,0])
plot(T,Y[:,1])
```
