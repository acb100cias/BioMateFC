```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta,n=3,2

x=np.linspace(0,3,200)

f=lambda x: beta/(1+x**n)

f3=f(f(f(x)))
```

```python
plt.plot(x,x)
plt.plot(x,f3)
plt.grid()
```

```python
import scipy.integrate as integrate
```

```python
def repre(x,t,beta,n):
  x1,x2,x3=x
  return np.array([f(x3)-x1,f(x1)-x2,f(x2)-x3])
```

```python
x0=np.array([1,1,1.2])
m=10000
beta=12
n=3
t=np.linspace(0,100,m)
sol=integrate.odeint(repre,x0,t,args=(beta,n))
```

```python
plt.plot(t,sol.T[0])
plt.plot(t,sol.T[1])
plt.plot(t,sol.T[2])
```
