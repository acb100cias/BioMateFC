```python
import matplotlib.pyplot as plt
import seaborn as sbn
import pandas as pd
import numpy as np
```

```python
def Die(CoZ):
    if isinstance(CoZ,Conejo) and CoZ.age>8.365:
        dr = 0.35
        return True if np.random.uniform() > 0.35 else False
```

```python
class Conejo:
    def __init__(self,gender=np.random.choice(['M','F']), age = 0, x = np.random.uniform(-10,10),y = np.random.uniform(-10,10)):
        self.x, self.y = x,y
        self.gender = gender
        self.age = age

    def move(self) -> None:
        m = 0.01
        self.x = self.x + np.random.uniform(-m,m)
        self.y = self.y + np.random.uniform(-m,m)

    def getOld(self) -> None:
        self.age += 1/365

    def __add__(self,Cp):
        if isinstance(Cp, Conejo):
            return Conejo(x=self.x,y=self.y)
```

```python
Conejos = [Conejo(x=np.random.uniform(-10,10),y=np.random.uniform(-10,10),gender=np.random.choice(['M','F'])) for i in range(50)]
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,8),dpi=120)
X, Y = [c.x for c in Conejos],[c.y for c in Conejos]
ax.scatter(X,Y,marker='*',color='red')
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,8),dpi=120)
for c in Conejos:
    c.move()

X, Y = [c.x for c in Conejos],[c.y for c in Conejos]
ax.scatter(X,Y,marker='*',color='red')
plt.show()
```
