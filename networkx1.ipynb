```python
import networkx as nw
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
G1=nw.Graph()
```

```python
G1.add_node(1)
G1.add_node(2)
```

```python
G1.add_edge(1,2)
```

```python
nw.draw(G1)
```

```python
N=list(range(100))
```

```python
E=[]
for i  in range(300):
  E.append((np.random.choice(N),np.random.choice(N)))
```

```python
G2=nw.Graph()
```

```python
G2.add_nodes_from(N)
G2.add_edges_from(E)
```

```python
nw.draw(G2)
```
