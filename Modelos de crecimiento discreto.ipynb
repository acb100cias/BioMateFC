# Ecuacion logística discreta

Consideremos ahora modelos de crecimiento discretos en tiempo. Es decir funciones de un conjunto a lo más numerable a  los reales. Tipicamente
$$
X:\mathbb{N}\rightarrow \mathbb{R}
$$
 Las ecuaciones en diferencias relacionana la tasa de cambio discreta (la equivalente discreta de la derivada) con alguna función. De la forma:
 
 $$
 \frac{X(t+1)-X(t)}{\Delta t}=F(X,t+1)
 $$
 
 Ahora, la dicerencia en $t$ es $1$ así que podemos utilizar lo anterior para dar una regla iterativa para $X(t+1)$. Veamos por ejemplo la versión discreta del crecimiento Malthusiano:
 
 $$
 \frac{X(t+1)-X(t)}{\Delta t}=\alpha X(t+1)
 $$
 de esto obtenemos:
 
 $$
 X(t+1)(1-\alpha)=X(t)
 $$
 
 finalmente
 $$
 X(t+1)=\frac{1}{(1-\alpha)}X(t)=G(t)
 $$
 
 Esta última expresión nos permite conocer el valor de la población para cualquier tiempo $t$, aplicando iteraciones de $G$, partiendo de un valor inicial $X_0=X(0)$:
 
 $$
 X(0)=X_0\\
 X(t+1)=G^t(X_0)
 $$

## Simulación numérica del Malthus discreto


```python
from numpy import *
from matplotlib.pyplot import *
```

```python
t=range(100)
```

```python
def f(s):
    return 1/(1-a)*s
```

```python
X=[0.1]
a=0.08
s0=X[0]
for i in t[1:]:
    X.append(f(X[-1]))
```

```python
plot(t,X)
plot(t,s0*exp(a*array(t)))
```

## Modelo logístico discreto

Ahora veamos la versión analoga del modelo logístico de crecimiento

$$
\frac{X(t+1)-X(t)}{\Delta t}=\alpha X(t+1)(1-\frac{X(t+1)}{K})
$$

```python
def g(s):
    return a*s*(1-s)+s
```

```python
X1=[0.1]
```

```python
for i in range(99):
    X1.append(g(X1[-1]))
```

```python
plot(t,X1)
```

```python
t=range(100)
for x0 in arange (0.1,2,0.2):
    X2=[x0]
    for i in t[1:]:
        X2.append(g(X2[-1]))
    plot(t,X2)
```
